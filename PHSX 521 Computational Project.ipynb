```python
#Claire Cook - Computational Project
#PHSX 521

#Parts a & b in this cell!
import numpy as np
import matplotlib.pyplot as plt


#################
#Define constants
R = 5 
g = 9.8
m = 1
#Define initial conditions: 
#initial position (φ_0), initial angular velocity (dφ/dt_0), and initial force on phi component:
phi0 = 0.34906585
phidot0 = 0.0
t0 = 0.0
#F0 = -m*g*np.sin(phi0)
F0 = (-g/R) * np.sin(phi0)
dt = 0.01 #time step 
tf = 10.0 #final time

#Initialize parameter arrays for plotting purposes:
phi_arr = []
phidot_arr = []
t_arr = []

#Start all parameters at initial conditions:
t=t0
phi = phi0
phidot = phidot0
F = F0
#Loop from initial time to final time in Verdet algorithm
while t<=tf:
    phi_tpdt = phi + (dt*phidot) + ((dt**2)/(2*m)) * (F) #φ(t+dt)
    F_tpdt = -(g/R)*np.sin(phi_tpdt) #F(t+dt)
    phidot_tpdt = phidot + (dt/(2*m))*(F_tpdt+F) #dφ/dt(t+dt)
    #update all parameters
    phi = phi_tpdt  
    F = F_tpdt
    phidot = phidot_tpdt
    #save parameter values to arrays
    t_arr.append(t)
    phi_arr.append(phi)
    phidot_arr.append(phidot)
    t+=dt #step time up 

#Plots phi and dphi/dt
plt.plot(t_arr, phi_arr, 'b-', label='Verdet') #phi
#plt.vlines(1.1, ymin=-phi0, ymax=phi0)
plt.xlabel('Time [s]')
plt.ylabel('Phi [radians]')
plt.title('Phi Plot')
#plt.show()

#Approximated solution for phi_0:
omega = np.sqrt(g/R)
t_arr = np.asarray(t_arr)
small_angle_approx_phi = phi0*np.cos(omega*t_arr)
plt.plot(t_arr, small_angle_approx_phi, 'k--', label='Approximation')
plt.legend()
plt.show()

#Clearly, looking at the graphs, this is a good approximation of the actual solution. 

#This plots the angular velocity over time, which was useful for my checking that my results made sense
plt.plot(t_arr, phidot_arr, 'r-') #phi dot
plt.ylabel('Angular Velocity [rad/s]')
plt.xlabel('Time [s]')
plt.title('Angular Velocity Plot')
plt.show()

```
