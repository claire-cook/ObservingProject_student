It also looks like they might have the 3727$\mathring A$ OII line in common. 

**(b)** Explain why they are in emission in one spectrum and in absorption in the other.  Use your knowledge of Kirkhoffs laws.

M57 is an emission spectrum because we're seeing the gas of the planetary nebula (particularly the hydrogen) emit at specific wavelengths corresponding to those elements/molecules' transitions. This is Kirchoff's law that a hot gas under a low pressure produces an emission line spectrum.

By contrast, we see absorption lines in Vega's spectrum because the light Vega emits interacts with the different atoms and molecules present in the star at the wavelengths corresponding to those elements, causing dips in flux at those wavelengths. This is Kirchoff's laws that a hot material under high pressure (like a star) gives off a continuous spectrum and that absorption lines are seen when a source of a continuous spectrum is observed behind cool gas under pressure.

Do your best given your knowledge of how light propogates through media to describe why Jupiter's spectrum is significantly redder than that of the sun.  
 

*As an aside, note that the very wide absorption lines longward of 6000$\mathring A$ in Jupiter's spectrum are absorption bands from molecules in Jupiter's atmosphere.  These bands absorb light over a wide wavelength range.*

**answer here**

Jupiter's spectrum is redder than the Sun's because Jupiter does not act as a perfect mirror reflection of the Sun. Although all of Jupiter's visible light is from reflected sunlight, when we look at Jupiter, it has distinct colors of its own due to its chemical composition. Similarly to how leaves are green because they reflect green light, Jupiter is redder than the sun because it reflects more red light.
